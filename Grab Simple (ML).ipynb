```python
#Removal of duplicated rows in labels dataframe
labels_0 = pd.read_csv("C:/Users/tanji/Desktop/safety/labels/part-00000-e9445087-aa0a-433b-a7f6-7f4c19d78ad6-c000.csv")
labels_0 = labels_0.groupby('bookingID', group_keys=True).apply(lambda x: x.loc[x.label.idxmax()])
labels_0 = labels_0.rename_axis(None)
```

```python
frames = [file_0, file_1, file_2, file_3, file_4, file_5, file_6, file_7, file_8, file_9]
main_df = pd.concat(frames)
```

```python
main_df.head()
```

```python
final_df = main_df.groupby('bookingID')
final_df = final_df['acceleration_x', 
                    'acceleration_y', 
                    'acceleration_z',
                    'gyro_x', 
                    'gyro_y',
                   'gyro_z'].agg([np.sum, np.mean, np.std])
```

```python
final_df.columns = ['acceleration_x_sum',
                    'acceleration_x_mean',
                    'acceleration_x_std',
                    'acceleration_y_sum',
                    'acceleration_y_mean',
                    'acceleration_y_std',
                    'acceleration_z_sum',
                    'acceleration_z_mean',
                    'acceleration_z_std',
                    'gyro_x_sum',
                    'gyro_x_mean',
                    'gyro_x_std',
                    'gyro_y_sum',
                    'gyro_y_mean',
                    'gyro_y_std',
                    'gyro_z_sum',
                    'gyro_z_mean',
                    'gyro_z_std',
                   ]
final_df.head()
```

```python
final_df = pd.merge(final_df, labels_0, on='bookingID', how='left')
```

```python
final_df.head()
```

```python
X = final_df.iloc[:,1:19].values
y = final_df.iloc[:,19].values

train_df_x, test_df_x, train_df_y, test_df_y = train_test_split(X, y,test_size=0.3)
```

```python
# Instantiate model with 1000 decision trees
rf = RandomForestClassifier(n_estimators = 10, random_state = 42 )

rf.fit(train_df_x, train_df_y)
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_df_x)
predictions
  
```

```python
print(confusion_matrix(test_df_y,predictions))  
print(classification_report(test_df_y,predictions))  
print(accuracy_score(test_df_y, predictions))  
```
